### Next Investigate Large number of NaNs in Cabin Column

Of the 889 entries, there are 687 nans for the Cabin column. I imagine those in certain cabins are more likely to survive; cabins lower down in the ship may be further from lifeboats etc. But imputing here is not appropriate. There may be some significance in the missing cabin entries: perhaps only lower socioeconomic status entries have no cabin entry - we could convert this into a binary variable for those with and without cabin entries. It is possible to develop two models to deal with passengers with and without a cabin entry, although the smaller sample size would increase uncertainty in the results.  
  
Below we see if there is any statistical significance of a cabin entry or not

```python
t_df['Has_Cabin'] = t_df['Cabin'].notna().astype(int)

contingency_table = pd.crosstab(t_df['Has_Cabin'], t_df['Survived'])

chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f'Chi-squared Test p-value: {p}')
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Has_Cabin', y='Survived', data=t_df)
plt.title('Survival Rate by Cabin')
plt.xlabel('Has Cabin')
plt.ylabel('Survival Rate')
plt.show()
```

Survival Rate is clearly a useful feature, and there are a significant number of NaNs, so we leave this for now.

## Categorical: Cabin, Ticket, Name

The data comes with little to no explanation of what the ticket data means (nor the cabin data, although this is more self-explanatory). Here we investigate the hypothesis that families / parties and the cabins and tickets are associated in a categorical sense; those on the same ticket or in the same cabin, are more likely to have the same surname.  
  
If this is the case, then knowing that Mr Karl Howell Behr survived, might indicate that Mrs Behr survived. And therefore if someone from Cabin C85 survived, then any other occupents of cabin C85 may have also survived.

```python
t_df['Cabin'].value_counts()[0:10]
```

```python
t_df[t_df['Ticket']=='CA. 2343']
```

All those on ticket CA. 2343 have the surname 'Sage'.

```python
t_df[t_df['Cabin'].isin(['B96 B98', 'G6','C23 C25 C27'])]
```

It is evident that those in the same cabin are usually on the same ticket, and often have the same surname.  
  
To use this information, we could encode the cabin and ticket features along with the surname of each passenger. Ideally, other family / party members would be present in the test dataset and knowing that all of the Carters survived, any Carters present in the test dataset would also likely survive.  
  
In practice, many of the entries in the Cabin column are nans, so only the ticket and surnames can be correlated. Secondly, we saw that the greatest number of passengers sharing a cabin was 4, given the large dataset (891 originally), this correlation would likely be overshadowed by more general trends to do with age or gender. Using such an engineered feature could likely cause the model to overfit, and if in some cases this trend is not true, we could see incorrect predictions. Finally, this association was only shown that those in the same cabin tend to be on the same tickets, not that those on the same ticket are in the same cabin, nor that those of the same surname are on the same ticket.  
  
In all, due to the small party size, presense of cabin nans, and absense of strong correlation, we will drop name, ticket and cabin. One final point for information present in Name: although this may already be represented in Pclass (which effectively indicates wealth), those like William Thornthon Carter the second are likely of a certain class. Given this, it may be possible to profile certain names and use these to predict survival.

```python
# Select columns to encode
columns_to_encode = ['Embarked', 'Pclass']

# Initialize the OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)  # drop='first' avoids multicollinearity by dropping one category

# Fit and transform the selected columns
encoded_features = encoder.fit_transform(t_df[columns_to_encode])

# Get the names of the new columns
encoded_feature_names = encoder.get_feature_names_out(columns_to_encode)

# Create a DataFrame for the encoded features
encoded_df = pd.DataFrame(encoded_features, columns=encoded_feature_names, index=t_df.index)

# Drop the original columns and add the encoded features
t_df = t_df.drop(columns=columns_to_encode).join(encoded_df)

# Display the updated DataFrame
t_df.head()
```

```python
y = t_df['Survived']
X = t_df.drop(columns=['Survived'])

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=14)
```

```python
# Create a pipeline with a scaler and a random forest classifier
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # StandardScaler to normalize features
    ('rf', RandomForestClassifier())  # Random Forest model with some parameters
])
```

```python
# Fit the pipeline to the training data
pipeline.fit(X_train, y_train)
```

```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('rf', RandomForestClassifier(random_state=42))
])

# Define the parameter grid to search
param_grid = {
    'rf__n_estimators': [50, 100, 200],       # Number of trees in the forest
    'rf__max_depth': [5, 10, 15, None],       # Maximum depth of the trees
    'rf__min_samples_split': [2, 5, 10],      # Minimum samples required to split a node
    'rf__min_samples_leaf': [1, 2, 4]         # Minimum samples required at each leaf node
}

# Set up GridSearchCV with cross-validation
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, 
                           cv=5, scoring='accuracy', n_jobs=-1, verbose=2)

# Fit GridSearchCV to the data
grid_search.fit(X_train, y_train)

# Print the best parameters and best score
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

```python
best_random_forest = grid_search.best_estimator_
```

```python
predictions=best_random_forest.predict(X_val)
```

```python
cm = confusion_matrix(y_val, predictions)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=best_random_forest.named_steps['rf'].classes_)
disp.plot(cmap='Blues')
```

```python
from sklearn.metrics import f1_score, accuracy_score

f1 = f1_score(y_val, predictions, average='weighted')
print(f"F1 Score: {f1:.2f}")


accuracy = accuracy_score(y_val, predictions)
print(f"Accuracy: {accuracy:.2f}")
```
